```python
# import necessary libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt

# read the satellite images for two different times
img1 = cv2.imread('dataset/5.jpg')
img2 = cv2.imread('dataset/3.jpg')

img1_resized = cv2.resize(img1, (img2.shape[1], img2.shape[0]))
img2_resized = img2

plt.imshow(img1_resized)
plt.show()

plt.imshow(img2_resized)
plt.show()


# convert the images to floats
img1_float = np.float32(img1_resized)
img2_float = np.float32(img2_resized)

# calculate NDVI for both images
ndvi1 = (img1_float[:, :, 2] - img1_float[:, :, 1]) / (img1_float[:, :, 2] + img1_float[:, :, 1])
ndvi2 = (img2_float[:, :, 2] - img2_float[:, :, 1]) /(img2_float[:, :, 2] + img2_float[:, :, 1])


# subtract the two NDVI images to identify changes
delta_ndvi = ndvi2 - ndvi1

# display the delta NDVI image
plt.imshow(delta_ndvi)
plt.colorbar()
plt.show()


```
