```python
# import necessary libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt

# read the satellite image
img = cv2.imread('dataset/1.jpg')

plt.imshow(img, cmap='viridis')
plt.colorbar()
plt.show()

# convert the image to float
img_float = np.float32(img)
# print(img_float)

# calculate NDVI
ndvi = (img_float[:, :, 2] - img_float[:, :, 1]) /(img_float[:, :, 2] + img_float[:, :, 1])

# display the NDVI image
plt.imshow(ndvi, cmap='viridis')
plt.colorbar()
plt.show()
```
