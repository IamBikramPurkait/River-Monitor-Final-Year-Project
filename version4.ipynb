```python
import os
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
import cv2
from rasterio.plot import show
from rasterio.mask import mask
# from osgeo import gdal



# img1 = cv2.imread('dataset/5.jpg')
# img2 = cv2.imread('dataset/3.jpg')

# img1_resized = cv2.resize(img1, (img2.shape[1], img2.shape[0]))
# img2_resized = img2

# cv2.imwrite("input1.jpg", img1_resized)
# cv2.imwrite("input2.jpg", img2_resized)


# Define file paths for Landsat images
image_path_1 = 'input1.jpg'
image_path_2 = 'input2.jpg'

# Open the images using rasterio
with rio.open(image_path_1) as src1, rio.open(image_path_2) as src2:
    # Read in the red and near-infrared bands for each image
    red_band1 = src1.read(2)
    nir_band1 = src1.read(3)
    red_band2 = src2.read(2)
    nir_band2 = src2.read(3)
    # Calculate NDWI for each image
    ndwi1 = (nir_band1.astype(float) - red_band1.astype(float)) / (nir_band1 + red_band1)
    ndwi2 = (nir_band2.astype(float) - red_band2.astype(float)) / (nir_band2 + red_band2)
    # Calculate the difference in NDWI between the two images
    ndwi_diff = ndwi2 - ndwi1
    # print(ndwi_diff)

# Plot the NDWI difference image
plt.imshow(ndwi_diff)
plt.colorbar()
plt.show()
```
